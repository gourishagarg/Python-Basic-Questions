13. What is the difference between for and while loops?
  - In Python, for and while loops are used to repeat things.

  - A for loop is used when you know how much times to repeat. For exaple, for i in range (5) will run 5 times.

  - A while loop is used when you dont know how many times, but want to repeat until a condition is True. For exaple, while i < 5 will keep running until i become 5.

  So, use for when th number is fixed, and while when it's not fixed.
